```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import squarify
import seaborn as sns
import plotly.graph_objects as go
import plotly.figure_factory as ff
import plotly.express as px
from plotly.subplots import make_subplots
from matplotlib.pylab import figure, linspace,sin, pi, subplots
from collections import defaultdict
```

```python
age_aggReports = pd.read_csv("../data/processed/exploded_data.csv",index_col=0)
age_aggReports[age_aggReports['patient_age'] > 0]
```

```python
required_outcomes = ['Death','Life Threatening','Hospitalization','Disability','Patient Visited ER']
```

```python
y = []
for outcome in required_outcomes:
    out_grp = age_aggReports[age_aggReports['patient_age'] > 0].groupby('category').get_group('Cosmetics').groupby('outcomes').get_group(outcome)
    y.append(out_grp['patient_age'])
fig1 = ff.create_distplot(y,required_outcomes,bin_size=5,show_hist=False,histnorm='probability',curve_type='kde')
fig1.update_layout(
    xaxis_title="Patient Age",
    yaxis_title="Probability Density",
    font=dict(
        family="Courier New, monospace",
        size=16,
        color="black",
        
    )
)
fig1.show()
```

```python
y = []
for outcome in required_outcomes:
    out_grp = age_aggReports[age_aggReports['patient_age'] > 0].groupby('category').get_group('Vit/Min/Prot/Unconv Diet(Human/Animal)').groupby('outcomes').get_group(outcome).dropna(subset=['patient_age'])
    y.append(out_grp['patient_age'])
fig1 = ff.create_distplot(y,required_outcomes,bin_size=5,show_hist=False,histnorm='probability',curve_type='kde')
fig1.update_layout(
    xaxis_title="Patient Age",
    yaxis_title="Probability Density",
    font=dict(
        family="Courier New, monospace",
        size=12,
        color="black"
    )
)
fig1.show()
```
